```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df1 = pd.read_csv("used_device_data.csv")
df1.head()
```

```python
print(df1['device_brand'].unique())           # For brands
print(df1['os'].unique())
```

```python
df1["device_age"] = 2025 - df1["release_year"]

df1
```

```python
df1["price_drop"] = df1["normalized_new_price"] - df1["normalized_used_price"]
df1
```

```python
df1.isnull().sum()
```

```python
df1['rear_camera_mp'] = df1.groupby('device_brand')['rear_camera_mp'].transform(lambda x: x.fillna(x.median()))
df1.isnull().sum()  # Check if nulls are handled
```

```python
df3 = df1.dropna()
df3.isnull().sum()
```

```python

df3.shape
```

```python
# df = pd.get_dummies(df3, columns=['device_brand', 'os','4g','5g'], drop_first=True)
# df
columnss=['device_brand', 'os','4g','5g']
from sklearn.preprocessing import OneHotEncoder
# Applying OneHotEncoder
encoder = OneHotEncoder(sparse_output=False,drop='first', handle_unknown='ignore')
encoded_features = encoder.fit_transform(df3[columnss])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(columnss))

# Drop Original Categorical Columns and Merge Encoded Data
df_encoded = df3.drop(columns=columnss).reset_index(drop=True)
df_encoded = pd.concat([df_encoded, encoded_df], axis=1)
df_encoded
```

```python
df = df_encoded.drop(['release_year','days_used'], axis=1)
df
```

```python
X = df.drop("normalized_used_price",axis='columns')
y = df.normalized_used_price
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)  # Returns a NumPy array
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor


# Import Models
from sklearn.linear_model import Ridge,Lasso

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Train each model separately
xgb_model = XGBRegressor(objective='reg:squarederror', n_estimators=80, max_depth=3, learning_rate=0.05)
xgb_model.fit(X_train, y_train)

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

ridge_model = Ridge(alpha=0.5)
ridge_model.fit(X_train, y_train)

lasso_model = Lasso(alpha=0.0001)
lasso_model.fit(X_train, y_train)

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

rf_model = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=15)
rf_model.fit(X_train, y_train)

# Function to evaluate a model
def evaluate_model(model, model_name):
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    print(f"📊 {model_name} Performance:")
    print(f"   R² Score: {r2:.4f}")
    print(f"   MAE: {mae:.4f}")
    print(f"   RMSE: {rmse:.4f}")
    print("-" * 40)

# Evaluate each model separately
evaluate_model(xgb_model, "XGBoost Regressor")
evaluate_model(linear_model, "Linear Regression")
evaluate_model(ridge_model, "Ridge Regression")
evaluate_model(lasso_model, "Lasso Regression")
evaluate_model(dt_model, "Decision Tree")
evaluate_model(rf_model, "Random Forest")


```
