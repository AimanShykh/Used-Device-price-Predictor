```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame with multiple numerical columns


# Function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        if df_clean[col].dtype in ['int64', 'float64']:  # Ensure numeric columns
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

    return df_clean


# Boxplot before removing outliers
plt.figure(figsize=(25, 10))
sns.boxplot(data=df3)
plt.title('Before Removing Outliers')
plt.show()


```

```python
# Remove outliers from all numerical columns
df_clean = remove_outliers_iqr(df3, df3.select_dtypes(include=['number']).columns)

```

```python
# Boxplot after removing outliers
plt.figure(figsize=(20, 10))
sns.boxplot(data=df_clean)
plt.title('After Removing Outliers')
plt.show()

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor


# Import Models
from sklearn.linear_model import Ridge,Lasso

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
def train_evaluate_model(model, model_name):
    pipeline = Pipeline([

        ("regressor", model)
    ])

    pipeline.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)

    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    print(f"{model_name} Performance:")
    print(f"  Train MSE: {train_mse:.4f} | Test MSE: {test_mse:.4f}")
    print(f"  Train RÂ² Score: {train_r2:.4f} | Test RÂ² Score: {test_r2:.4f}")
    print("-" * 50)

    return model, train_mse, test_mse, train_r2, test_r2
```

```python
xgb_model = XGBRegressor(objective='reg:squarederror', n_estimators=80, max_depth=3, learning_rate=0.05)
xgb=train_evaluate_model(xgb_model, "XGBoost Regressor")

linear_model = train_evaluate_model(LinearRegression(), "Linear Regression")
ridge_model = train_evaluate_model(Ridge(alpha=1.0), "Ridge Regression")
lasso_model = train_evaluate_model(Lasso(alpha=0.01), "Lasso Regression")
dt_model = train_evaluate_model(DecisionTreeRegressor(random_state=42), "Decision Tree")
rf_model = train_evaluate_model(RandomForestRegressor(n_estimators=100, random_state=42,max_depth=15), "Random Forest")
```
